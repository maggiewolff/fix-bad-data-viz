```python
# import data

help_others = pd.read_csv('help_others.csv')
obey = pd.read_csv('obey.csv')
popular = pd.read_csv('popular.csv')
think_for_self = pd.read_csv('think_for_self.csv')
work_hard = pd.read_csv('work_hard.csv')
```

### Prepare dataframe for visuals 

```python
# add column with topic name 

help_others['topic'] = 'help_others'
obey['topic'] = 'obey'
popular['topic'] = 'popular'
think_for_self['topic'] = 'think_for_self'
work_hard['topic'] = 'work_hard'
```

```python
# combine into one dataframe 
df = pd.concat([help_others, obey, popular, think_for_self, work_hard], ignore_index = True)

# preview new dataframe 
df
```

```python
# check the column names 
df.columns
```

```python
# drop X.1 column as it is all null values 
df = df.drop(columns=['X.1'])
```

```python
#rename "Year" column 
df = df.rename(columns={"Year": "importance"})
```

```python
# switch the year columns into rows 

df2 = pd.melt(df, id_vars=['importance', 'topic'], value_vars=['1986', '1987', '1988', '1989', '1990', '1991', '1993',
       '1994', '1996', '1998', '2000', '2002', '2004', '2006', '2008', '2010',
       '2012', '2014', '2016', '2018', '2021', '2022'],
        var_name='year', value_name='percent_of_responses')
```

```python
# view new dataframe 
df2
```

### Vizualize data

```python
for i in df2["topic"].unique(): # iterate through the topic names to create individual visuals 
    
    # create visual 
    ax = sns.lineplot(data=df2[df2["topic"]==i], x="year", y="percent_of_responses", hue="importance");
    
    # move the legend to outside of the visual 
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1));
    
    # rotate the x-axis labels so they do not overlap
    plt.xticks(rotation=90);
    
    # add a title using the topic name 
    plt.title(i);

    # show the visualization
    plt.show();
```

```python
# create visual comparing each topic by the percent selecting it as "most important" 
ax = sns.lineplot(data=df2[df2["importance"]=="MOST IMPORTANT"], x="year", y="percent_of_responses", hue="topic");
plt.xticks(rotation=90);
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1));
plt.title("MOST IMPORTANT");
plt.show();
```

```python
# create visual comparing each topic by the percent selecting it as "least important" 
ax = sns.lineplot(data=df2[df2["importance"]=="LEAST IMPORTANT"], x="year", y="percent_of_responses", hue="topic");
plt.xticks(rotation=90);
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1));
plt.title("LEAST IMPORTANT");
plt.show();
```
